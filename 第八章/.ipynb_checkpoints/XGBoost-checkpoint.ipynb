## XGboost实战

```python
import pandas as pd
from sklearn.metrics import accuracy_score
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import xgboost as xgb
```

### 1. 加载数据

```python
dtrain = xgb.DMatrix("./data/agaricus.txt.train")
dtest = xgb.DMatrix("./data/agaricus.txt.test")
```

### 2.设置参数

```python
param = {"max_depth": 3, "eta":1, "silent":0, "objective":"binary:logistic"}
num_round = 2 #设置迭代次数
```

### 3.训练预测

```python
bst = xgb.train(param,dtrain, num_round)
train_preds = bst.predict(dtrain)
train_preds
```

#### 训练集预测

```python
train_predictions = [round(value) for value in train_preds] #四舍五入，确定分类标签
y_train = dtrain.get_label()
print(y_train)
train_accuracy = accuracy_score(y_train, train_predictions)
print("Accuracy: %.2f"%(train_accuracy*100.0))
```

#### 测试集预测

```python
test_preds = bst.predict(dtest)
test_predictions = [round(value) for value in test_preds]
y_test = dtest.get_label()
test_accuracy = accuracy_score(y_test, test_predictions)
print("Accuracy: %.2f"%(test_accuracy*100.0))
```

### 4.模型可视化
